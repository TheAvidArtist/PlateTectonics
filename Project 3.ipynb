## Graphing time vs magnitude of earthquakes

```python
Data=pd.read_csv('query.csv')
Dataset = gv.Dataset(Data, kdims=['latitude', 'longitude', 'mag', 'depth', 'time'])
April2015=pd.read_csv('April2015.csv')
April = gv.Dataset(April2015, kdims=['latitude', 'longitude', 'mag', 'depth', 'time'])
Divergent=pd.read_csv('DivergentData.csv')
DivergentDataset = gv.Dataset(Divergent, kdims=['latitude', 'longitude', 'mag', 'depth', 'time'])
Transform=pd.read_csv('TransformData.csv')
TransformDataset = gv.Dataset(Transform, kdims=['latitude', 'longitude', 'mag', 'depth', 'time'])


mag = []
lat = []
long = []
time = []

for i in range (0, len(Data)):
    time.append(dateutil.parser.isoparse(Data['time'][i]))
    lat.append(Data['latitude'][i])
    long.append(Data['longitude'][i])
    mag.append(Data['mag'][i])
    
dates = matplotlib.dates.date2num(time)

mag1 = []
lat1 = []
long1 = []
time1 = []

for i in range (0, len(April2015)):
    time1.append(dateutil.parser.isoparse(April2015['time'][i]))
    lat1.append(April2015['latitude'][i])
    long1.append(April2015['longitude'][i])
    mag1.append(April2015['mag'][i])
    
dates1 = matplotlib.dates.date2num(time1)

mag2 = []
lat2 = []
long2 = []
time2 = []

for i in range (0, len(Divergent)):
    time2.append(dateutil.parser.isoparse(Divergent['time'][i]))
    lat2.append(Divergent['latitude'][i])
    long2.append(Divergent['longitude'][i])
    mag2.append(Divergent['mag'][i])
    
dates2 = matplotlib.dates.date2num(time2)

mag3 = []
lat3 = []
long3 = []
time3 = []

for i in range (0, len(Transform)):
    time3.append(dateutil.parser.isoparse(Transform['time'][i]))
    lat3.append(Transform['latitude'][i])
    long3.append(Transform['longitude'][i])
    mag3.append(Transform['mag'][i])
    
dates3 = matplotlib.dates.date2num(time3)


plt.figure(1)
plt.plot_date(dates, mag)
plt.figure(1).set_figheight(10)
plt.figure(1).set_figwidth(30)
plt.title('time vs magnitude')
plt.xlabel('time')
plt.ylabel('magnitude of earthquake')

plt.figure(2)
plt.figure(2).set_figheight(10)
plt.figure(2).set_figwidth(30)
plt.plot_date(dates1, mag1)
plt.title('time vs magnitude of April 2015')
plt.xlabel('time')
plt.ylabel('magnitude of earthquake')

plt.figure(3)
plt.figure(3).set_figheight(10)
plt.figure(3).set_figwidth(30)
plt.plot_date(dates2, mag2)
plt.title('time vs magnitude of Divergent Boundary')
plt.xlabel('time')
plt.ylabel('magnitude of earthquake')

plt.figure(4)
plt.figure(4).set_figheight(10)
plt.figure(4).set_figwidth(30)
plt.plot_date(dates3, mag3)
plt.title('time vs magnitude of Transform Boundary')
plt.xlabel('time')
plt.ylabel('magnitude of earthquake')
```

```python
def average(num):
    sum_num = 0
    for t in range(0, len(num)):
        sum_num = sum_num + num[t]           

    avg = sum_num / len(num)
    return avg
```

```python
MaxData=max(Data['mag'])
MinData=min(Data['mag'])
AvgData=average(Data['mag'])

MaxDivData=max(Divergent['mag'])
MinDivData=min(Divergent['mag'])
AvgDivData=average(Divergent['mag'])

MaxTransData=max(Transform['mag'])
MinTransData=min(Transform['mag'])
AvgTransData=average(Transform['mag'])

print('Max overall earthquake: ', MaxData)
print('Min overall earthquake: ', MinData)
print('Average earthquakes overall: ', AvgData)

print('Max Divergent boundary earthquake: ', MaxDivData)
print('Min Divergent boundary earthquake: ', MinDivData)
print('Average Divergent boundary earthquakes: ', AvgDivData)

print('Max Transform boundary earthquake: ', MaxTransData)
print('Min Transform boundary earthquake: ', MinTransData)
print('Average Transform boundary earthquakes: ', AvgTransData)
```
